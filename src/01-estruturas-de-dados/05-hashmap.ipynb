# hashmap

coleção de chaves e valores.

não importa quantos itens tiver, sempre que precisar acessar uma chave, a complexidade é O(1).  
agora, como que as entradas são possíveis ser encontradas dentro do array em tempo constante?

a mágica é a **hash function**, um dos componentes principais de um hashmap.

a maior parte das implementações de hashmaps usam um (ou mais) array(s) para salvar em memória os dados do hashmap.  
sabendo disso, a função da **hash function** é de mapear as chaves para o memória em si que cada chave aponta.
